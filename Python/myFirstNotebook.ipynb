```python
import pandas as pd
import numpy as np

from sklearn.datasets import fetch_openml
```

```python
titanic = fetch_openml(name='titanic', version=1, as_frame=True)
df = titanic.frame
df.head()
```

```python
# Check for missing values and handle them
df.info()
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
# Handling the missing values
median_age = df['age'].median()
df['age'].fillna(median_age, inplace=True)
```

```python
df.isnull().sum()
```

```python
# 2. Check for duplicates and removing them
df.duplicated().sum()
```

```python
# 3. 

df.sex.value_counts()

df['sex'] = df.sex.str.upper()

df.sex.value_counts()
```

```python
# 4. Detect outliers in the Age and Fare cols
# OUTLIERS > 1.5*IQR

Q1_age = df['age'].quantile(0.25)
Q3_age = df['age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

Q1_fare = df['fare'].quantile(0.25)
Q3_fare = df['fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare

print(IQR_age, IQR_fare)
```

```python
age_outliers = df[(df['age'] < (Q1_age - 1.5*IQR_age)) | (df['age'] > (Q3_age + 1.5*IQR_age))]
fare_outliers = df[(df['fare'] < (Q1_fare - 1.5*IQR_fare)) | (df['fare'] > (Q3_fare + 1.5*IQR_fare))]
```

```python
# Seeing the above
fare_outliers.head()
```

```python
# 5. Create a new "Family Size" based on "SibSp" and "Parch".
df['family_size'] = df['sibsp'] + df['parch'] + 1
df.head()
```

```python
# 6. Encode the embarked col
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['embarked_encoded'] = label_encoder.fit_transform(df['embarked'])
df.head()
```

```python
df['embarked_encoded'].value_counts()
```

```python
df['embarked'].unique()
```

```python
# One Hot Encoding

df['embarked_S'] = df.apply(
    lambda row: 1 if (row['embarked']=='S') else 0,
    axis=1
)

df['embarked_C'] = df.apply(
    lambda row: 1 if (row['embarked']=='C') else 0,
    axis=1
)

df['embarked_Q'] = df.apply(
    lambda row: 1 if (row['embarked']=='Q') else 0,
    axis=1
)
df.head(n=10)
```

```python
# 8. Clean the Name 
import re

df['name'] = df['name'].apply(
    lambda x: re.sub(r'(Mr\.|Mrs\.|Dr\.|Miss\.|Master\.)', '', x) 
)

df['name'] = df.name.str.lower()

df.head(n=10)
```
